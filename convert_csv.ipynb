```python
import csv


def utf2sjis(input_file):
    #input_file='test.csv'
    output_file=input_file.replace('.csv','')+'_sjis.csv'

    print(output_file)

    with open(input_file, encoding='utf8') as f_in:
        with open(output_file, 'w', encoding='cp932') as f_out:
            f_out.write(f_in.read())
```

```python
import glob

files=glob.glob('./data/**/*')

for file in files:
    print(file)
    if(file not in 'sjis'):
        utf2sjis(file)


```
